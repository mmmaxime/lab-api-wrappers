Function to go through all the playlist items of the playlist page

```python
from random import randint
from time import sleep

def get_playlist_tracks(playlist_id):
    playlist = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = playlist['items']
    while playlist['next']!=None:          # while there is still a next ply to go to
        playlist = sp.next(playlist)
        tracks = tracks + playlist['items']
        sleep(randint(1,2))
    return tracks
```

### LAB 6.07 Unsupervised learning 

Separating the numerical and categorical data to later scale the numerical data for the Elbow and Silhouette methods

```python
num_data = flat_data_clean1.select_dtypes(include = np.number)
cat_data = flat_data_clean1.select_dtypes(include = object)
num_data
```

Scaling the data using the Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
#X_prep = StandardScaler().fit_transform(num_data)
scaler = StandardScaler().fit(num_data)
X_prep = scaler.transform(num_data)

# Now, all features will have the same weight.
pd.DataFrame(X_prep,columns=num_data.columns).head()
```

Elbow Method

```python
from sklearn.cluster import KMeans
K = range(2, 10)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters = k,
                    random_state = 1234)
    kmeans.fit(X_prep)
    inertia.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

Silhouette Method

```python
from sklearn.metrics import silhouette_score
K = range(2, 10)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters = k,
                    random_state = 1234)
    kmeans.fit(X_prep)
    silhouette.append(silhouette_score(X_prep, kmeans.predict(X_prep)))
```

```python
plt.figure(figsize=(12,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Slhouette score showing the optimal k')
```

KMeans Method

```python
from sklearn import cluster
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=8, random_state=1234)
kmeans.fit(X_prep)
```

```python
import pandas as pd
clusters = kmeans.predict(X_prep)
pd.Series(clusters).value_counts().sort_index()
```

```python
title = input("Enter a song you like: ").lower()
artist = input("Enter an artist you like: ").lower()

if Top100['song'].str.lower().str.contains(str(title).lower()).any() == True:
        print('here what I found for you! ')
        print(Top100['song'].sample())
else:
    # get the track name
    results = sp.search(q="track:" + title, type="track")
    
    # find the song uri with spotipy
    song_uri = results['tracks']['items'][0]['uri'].split(':')[2]
    
    #find song features
    song_features = sp.audio_features(song_uri)
    
    #clean song features
    new = pd.DataFrame.from_dict(song_features)
    track_href = new[['track_href']]
    X_pred = new.drop(columns=['type' ,'id','track_href','uri','analysis_url','mode','key','mode','time_signature'])
    
    # scale the data
    X_scaled = scaler.transform(X_pred)
    X_scaled = pd.DataFrame(X_scaled,columns=X_pred.columns)
    
    # use the Kmeans created before to predict the cluster for the song
    look_for = kmeans.predict(X_scaled)
    print('Here what I found for you !!!')
    
    # choose a random row from the platlist with same cluster
    a = (final1.loc[final1.clusters == look_for[0],['titles','artists','id']].sample(1))
    #print(final1.loc[final1.clusters == look_for[0],['titles','artists','id']].sample(1))
    print(a.iloc[0,2])
    print('Titles:', a.iloc[0,0] ,'Artists:', a.iloc[0,1],'URL:',"https://open.spotify.com/track/" + a.iloc[0,2] + "?si=b7b938ef36254e36&nd=1")
```
